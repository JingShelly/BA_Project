```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download SPY data
spy = yf.download('SPY', end='2025-03-31')  # Ensure the 'end' date format is correct
spy = spy[['Close']]  # We're only interested in the closing prices

# Splitting the dataset into train, validation, and test sets
train_size = int(len(spy) * 0.8)
test_size = 30
val_size =  len(spy) - train_size - test_size


train_set = spy[:train_size]
val_set = spy[train_size: train_size + val_size]
test_set = spy[train_size + val_size:]

# Plotting the data
plt.figure(figsize=(14, 6))
plt.plot(train_set.index, train_set['Close'], label='Training dataset', color='blue')
plt.plot(val_set.index, val_set['Close'], label='Validation dataset', color='orange')
plt.plot(test_set.index, test_set['Close'], label='Test dataset', color='red')
plt.title('SPY Closing Prices - Train, Validation, Test Split')
plt.xlabel('Date')
plt.ylabel('SPY Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
